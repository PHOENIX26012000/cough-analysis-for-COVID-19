# Mel-spectrogram of Cough Signals

```python
audio_path = "C:/Users/digvi/Anaconda3/ESC-50/audio/audio/1-19111-A-24.wav"
y, sr = librosa.load(audio_path)
S = librosa.feature.melspectrogram(y=y, sr=sr,n_mels=128, fmax=8000)
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(S, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram of Cough Signals')
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
librosa.display.waveplot(y=y,sr=sr)
```

```python
audio_path = "C:/Users/digvi/Anaconda3/ESC-50/audio/audio/1-19118-A-24.wav"
y1, sr1 = librosa.load(audio_path)
S = librosa.feature.melspectrogram(y=y1, sr=sr1,n_mels=128, fmax=8000)

plt.figure(figsize=(10,4))
librosa.display.specshow(librosa.power_to_db(S, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram of Cough Signals')
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
librosa.display.waveplot(y=y1, sr=sr1)
```

# MFCC

```python
a = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=500)
plt.figure(figsize=(10,5))
librosa.display.specshow(a, x_axis='time')
plt.colorbar()
plt.title('MFCC of Cough Signal')
plt.tight_layout()
plt.show()
```
